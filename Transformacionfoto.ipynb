```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Cargar imagen RGB
imagen = Image.open("enfermeros1.jpg").convert("RGB")
matriz_rgb = np.array(imagen)

# Escala de grises
imagen_gris = imagen.convert("L")
matriz_gris = np.array(imagen_gris)

# Invertir colores
invertida = 255 - matriz_rgb

# Detección de bordes (simple por diferencia de píxeles verticales)
bordes = np.abs(np.diff(matriz_gris, axis=0))

# Visualización
fig, axs = plt.subplots(1, 4, figsize=(20, 5))

axs[0].imshow(matriz_rgb)
axs[0].set_title("Original RGB")
axs[0].axis("off")

axs[1].imshow(matriz_gris, cmap="gray")
axs[1].set_title("Escala de grises")
axs[1].axis("off")

axs[2].imshow(invertida)
axs[2].set_title("Colores invertidos")
axs[2].axis("off")

axs[3].imshow(bordes, cmap="gray")
axs[3].set_title("Detección de bordes")
axs[3].axis("off")

plt.tight_layout()
plt.show()
```
