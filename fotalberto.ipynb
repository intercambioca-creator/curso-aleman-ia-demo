```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Cargar imagen en color
imagen = Image.open("Alberto_digital.JPG").convert("RGB")
matriz_rgb = np.array(imagen)

# Mostrar imagen original
plt.imshow(matriz_rgb)
plt.title("Imagen original (RGB)")
plt.axis("off")
plt.show()

# Mostrar dimensiones
print("Dimensiones:", matriz_rgb.shape)
print("Primeros p√≠xeles:", matriz_rgb[:2, :2])
```

```python
# Separar canales
rojo = matriz_rgb[:, :, 0]
verde = matriz_rgb[:, :, 1]
azul = matriz_rgb[:, :, 2]

# Mostrar canales
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

axs[0].imshow(rojo, cmap="Reds")
axs[0].set_title("Canal Rojo")
axs[0].axis("off")

axs[1].imshow(verde, cmap="Greens")
axs[1].set_title("Canal Verde")
axs[1].axis("off")

axs[2].imshow(azul, cmap="Blues")
axs[2].set_title("Canal Azul")
axs[2].axis("off")

plt.show()
```
