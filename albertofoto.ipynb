```python
from PIL import Image
import numpy as np

# Cargar la imagen en color
imagen = Image.open("Foto_Alberto_digital.JPG").convert("RGB")

# Convertir a matriz NumPy
matriz_rgb = np.array(imagen)

# Mostrar dimensiones: alto x ancho x canales
print("Dimensiones de la imagen:", matriz_rgb.shape)

# Mostrar los primeros 2x2 p√≠xeles (cada uno con 3 valores RGB)
print("Fragmento de la matriz RGB:")
print(matriz_rgb[:2, :2])
```

```python
import matplotlib.pyplot as plt

# Mostrar la imagen original
plt.figure(figsize=(6, 6))
plt.imshow(matriz_rgb)
plt.title("Imagen original (RGB)")
plt.axis("off")
plt.show()

# Mostrar cada canal por separado
canal_rojo = matriz_rgb[:, :, 0]
canal_verde = matriz_rgb[:, :, 1]
canal_azul = matriz_rgb[:, :, 2]

# Visualizar los canales
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

axs[0].imshow(canal_rojo, cmap="Reds")
axs[0].set_title("Canal Rojo")
axs[0].axis("off")

axs[1].imshow(canal_verde, cmap="Greens")
axs[1].set_title("Canal Verde")
axs[1].axis("off")

axs[2].imshow(canal_azul, cmap="Blues")
axs[2].set_title("Canal Azul")
axs[2].axis("off")

plt.show()
```
